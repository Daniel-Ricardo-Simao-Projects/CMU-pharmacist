```python
import json
from unidecode import unidecode



# open the json file
with open('pharmacy.json', 'r') as f:
    data = f.read()
    data = json.loads(data)
    for pharmacy in data['results']:
        name = pharmacy['name']
        address = pharmacy['vicinity']
        image_name = unidecode(name.replace(" ", "").replace(".", ""))
        print(f'(\"{name}\", \"{address}\", \"internal/images/pharmacies/{image_name}.png\"),')

with open('pharmacy2.json', 'r', encoding='utf-8') as f:
    data = f.read()
    data = json.loads(data)
    for pharmacy in data['results']:
        name = pharmacy['name']
        address = pharmacy['vicinity']
        image_name = unidecode(name.replace(" ", "").replace(".", ""))
        print(f'(\"{name}\", \"{address}\", \"internal/images/pharmacies/{image_name}.png\"),')
```
